```python
from letta_client import CreateBlock, Letta, MessageCreate, AsyncLetta
client = AsyncLetta(
    base_url="http://localhost:8283",
)
```

```python
agent_id = "agent-d5bbe3d9-6f4a-496c-a455-9b6ef4b82b5d"
await client.agents.reset_messages(agent_id=agent_id)
```

```python
stream = client.agents.messages.create_stream(
    agent_id=agent_id,
    messages=[
        MessageCreate(
            role="user",
            content="Hello, can you explain the benefits of the card for me?",
        ),
    ],
    stream_tokens=True,
)

num_to_skip = 12
characters = 0
last_chunk = ""
async for chunk in stream:
    if chunk.message_type == "assistant_message":
        # Check for quote-comma sequence in current chunk or spanning across chunks
        if "\",\"" in chunk.content or (last_chunk and (last_chunk.endswith("\"") and chunk.content.startswith(",\"") or last_chunk.endswith(",") and chunk.content.startswith("\""))):
            break
            
        if characters < num_to_skip:
            if len(chunk.content) <= num_to_skip - characters:
                characters += len(chunk.content)
                continue
            else:
                print(chunk.content[(num_to_skip - characters):], end="", flush=True)
                characters = num_to_skip
        else:
            print(chunk.content, end="", flush=True)
        
        last_chunk = chunk.content
```
